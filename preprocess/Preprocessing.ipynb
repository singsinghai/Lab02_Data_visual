# Chuẩn hóa kiểu dữ liệu của các cột

## Chuẩn hóa cột `Date` về dạng datetime

```python
data = df.copy()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data
```

## Chuẩn hóa các cột numeric

```python
excludes = ['Country', 'Continent', 'Date']#, 'New Deaths', '1 Test every X ppl', 'New Deaths/1M pop ']
numeric = data.loc[:, ~data.columns.isin(excludes)]
```

```python
def replace_empty(x):
    if x == '' or x == ' ':
        x = 'NaN'
    return x

def to_number(s):
    #print(s)
    s = s.astype('str')
    s = s.str.replace(',', '')
    s = s.apply(replace_empty)
    s = s.astype('float')
    return s

normalized_numeric = numeric.apply(to_number)
```

```python
for col in numeric.columns:
    data[col] = normalized_numeric[col]
    
data
```

## Thêm cột Country Codes

Mã quốc gia được lấy từ https://pypi.org/project/country-converter/

Dọn sạch bằng phương pháp strip

```python
def clean_space_in_country_col(country):
    return country.strip()
        
data['Country'] = data['Country'].apply(clean_space_in_country_col)
```

```python
to_country_code = json.loads(open('country-codes.json','r').read())['Country Code']

def create_country_code(x):
    try:
        return to_country_code[x]
    except:
        return ''

data['Country Code'] = data['Country'].apply(create_country_code)
```
